## Query CosmosDB
First we need to connect to our Cosmos instance, we store the credentials in 'cosmoskeys.json' located in the same folder.

```python
import azure.cosmos.cosmos_client as cosmos_client
import json
from datetime import datetime
with open('./cosmoskeys.json') as f:
    keys = json.load(f)
    url = keys['cosmosurl']
    key = keys['cosmoskey']
    datapane_key = keys['datapane']
```

```python
client = cosmos_client.CosmosClient(url, {'masterKey': key})
database = client.get_database_client('arduino')
```

## Select all readings
Select all the temperature readings we've logged so far.

```python
temps_container = database.get_container_client('temps')

query = "SELECT * FROM c"

readings = list(temps_container.query_items(
    query=query,
    enable_cross_partition_query=True
))

request_charge = dashboard_container.client_connection.last_response_headers['x-ms-request-charge']

print('Query returned {0} items. Operation consumed {1} request units'.format(len(readings), request_charge))
```

# Data preparation
Load the data into a DataFrame and preprocess it for some basic charting.

```python
import pandas as pd

df = pd.DataFrame.from_dict(readings)

df = df.dropna()

df.head()
```

```python
#timestamp preparation the hard way because we'll reuse this logic later without Pandas
datetime.fromtimestamp(df.iloc[0]['timestamp']).strftime("%Y-%m-%d-%H")
```

```python
#adding a new column with the prepared date
df['time_stamp'] = df['timestamp'].apply(lambda x: 
                          datetime.fromtimestamp(x).strftime("%Y-%m-%d-%H"))
df['time_stamp'] = pd.to_datetime(df['time_stamp'])
```

```python
df.head()
```

```python
#ok so a basic chart, pretty ugly because of all the holes in the data
aggregates = df[['temp','time_stamp']].groupby(['time_stamp']).mean()
aggregates.plot(figsize=(15,9))
```

# Building the Aggregates
Let's iterate over records, group by hour and add a flag to say we've already processed the record so we know to skip it in the future, then write back the aggregates to Cosmos.

```python
counts = {}
for i in readings:
    payload = i
    dt_object = datetime.fromtimestamp(i['timestamp'])
    dt_d = dt_object.strftime("%Y-%m-%d:%H")
    if dt_d in counts.keys():
        counts[dt_d]={'temp_sum':counts[dt_d]['temp_sum']+i['temp'],'count':counts[dt_d]['count']+1,
                     'illum_sum':counts[dt_d]['illum_sum']+i['illuminance']}
    else:
        counts[dt_d]={'temp_sum':i['temp'],'count':1,
                     'illum_sum':i['illuminance']}
    
```

```python
counts['2020-12-09:16']
```

```python
#initialize our container for aggregations
dashboard_container = database.get_container_client('dashboard')

#iterate through all our readings, parse the timestamp, build a record with aggregated readings per hour
counts = {}
for i in readings:
    payload = i
    dt_object = datetime.fromtimestamp(i['timestamp'])
    dt_d = dt_object.strftime("%Y-%m-%d:%H")
    #do we already have a record in the dict for this hour?
    if dt_d in counts.keys():
        counts[dt_d]={'temp_sum':counts[dt_d]['temp_sum']+i['temp'],'count':counts[dt_d]['count']+1,
                     'illum_sum':counts[dt_d]['illum_sum']+i['illuminance']}
    else:
        counts[dt_d]={'temp_sum':i['temp'],'count':1,
                     'illum_sum':i['illuminance']}
    #modify the record to add a flag that we have processed it
    temps_container.upsert_item(
                {
                    'id': 'ard1010wifi'+str(payload['timestamp']),
                    'temp': payload['temp'],
                    'timestamp': payload['timestamp'],
                    'humidity': payload['humidity'],
                    'pressure': payload['pressure'],
                    'illuminance' : payload['illuminance'],
                    'date':dt_object.strftime("%Y-%m-%d"),
                    'processed':True
                })
```

```python
#write the aggregates out to our dashboarding container
for i in counts:
    print(i,counts[i])
    dashboard_container.upsert_item(
    {
        'id':i,
        'date':i,
        'temp_sum':counts[i]['temp_sum'],
        'count':counts[i]['count'],
        'illum_sum':counts[i]['illum_sum']
    })
```

```python
#now how many readings do we have for aggregation?
query = "SELECT * FROM c"

counts = list(dashboard_container.query_items(
    query=query,
    enable_cross_partition_query=True
))

request_charge = dashboard_container.client_connection.last_response_headers['x-ms-request-charge']

print('Query returned {0} items. Operation consumed {1} request units'.format(len(counts), request_charge))
```

```python
#how many records do we have that have not been processed?
#note: we can't look for processed is not true, for unprocessed records they lack the attribute entirely
query = "SELECT * FROM c WHERE NOT is_defined(c.processed)"

items = list(temps_container.query_items(
    query=query,
    enable_cross_partition_query=True
))

request_charge = temps_container.client_connection.last_response_headers['x-ms-request-charge']

print('Query returned {0} items. Operation consumed {1} request units'.format(len(items), request_charge))
```
